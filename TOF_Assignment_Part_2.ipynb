```python
import pandas as pd

df = pd.read_csv("AB_NYC_2019.csv",index_col="neighbourhood_group")
```

```python
import matplotlib.pyplot as plt
```

```python
no_of_airbnbs=df[['id','host_id','name','host_name','neighbourhood']]
df1=no_of_airbnbs.groupby("neighbourhood_group").count()
df1.plot.bar()
```

    As we see from above from all the five neighbourhood groups the highest number of airbnbs are in Manhattan and lowest number is in Staten Island

```python
price=df['price']
df2=price.groupby("neighbourhood_group").sum()/10**5
plt.ylabel('Total Prices')
df2.plot.bar()
```

From above bar graph, we can see the Total Prices corresponding to a neighbourhood group. Here Manhattan has highest and Staten Island has low total price of all their Airbnbs 

```python
hosts=df['calculated_host_listings_count']
df2=hosts.groupby("neighbourhood_group").sum()/1000
plt.ylabel('Total hosts')
df2.plot.bar()
```

The above graph shows us the total number of hosts corresponding to a particular neighbourhood group

```python
hosts=df[df['calculated_host_listings_count']>100]
plt.bar(hosts['host_name'],hosts['calculated_host_listings_count'])
plt.xlabel("Host name")
plt.ylabel("Calculated host listings")
plt.show()
```

If we define business of a host based on the number of listings that host has then the above hosts are the busiest since they all have more than 100 listings to manage.

```python
df = pd.read_csv("AB_NYC_2019.csv")
```

```python
df1=df[df['reviews_per_month']>10]
plt.bar(df1['neighbourhood_group'],df1['reviews_per_month'])
plt.xlabel('Neighbourhood group')
plt.ylabel("Reviews per month")
plt.show()
```

There does exist a noticable difference between traffic in neighbourhood groups as shown in the above graph, here if we define the traffic in a particular neighbourhood group as the total number of reviews per month all the listings in that neighbourhood group has then Manhattan has significantly higher number of reviews per month than other neighbourhood groups.

In all of the above graphs we have seen use of various columns like neighbourhood_group,reviews_per_month,prices etc and we can valuable insights from all of those. We can have seen that Manhattan is ahead of all Neighbourhood groups in every aspect and at the same time Staten Island is lagging at last position in same aspects.
